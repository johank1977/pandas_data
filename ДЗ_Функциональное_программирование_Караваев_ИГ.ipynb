# Задача 1. Фильтрация и трансформация данных (lambda, map, filter)
Дан список словарей, представляющих данные о студентах:

```python
students = [
{'name': 'Алексей', 'grade': 85, 'city': 'Москва'},
{'name': 'Мария', 'grade': 92, 'city': 'Казань'},
{'name': 'Иван', 'grade': 78, 'city': 'Москва'},
{'name': 'Елена', 'grade': 88, 'city': 'Иннополис'},
{'name': 'Дмитрий', 'grade': 55, 'city': 'Москва'},
]
```

Используя filter и lambda, отфильтруйте студентов, чья оценка (grade) выше 80.

```python
filtered_list = list(filter(lambda d: d['grade'] > 80, students))  
print(filtered_list)  
```

Используя map и lambda, преобразуйте отфильтрованный список, оставив только имена студентов в верхнем регистре.

```python
uppercase_words = list(map(lambda d: d['name'].upper(), filtered_list))
print(uppercase_words)
```

Используя reduce (из functools) и lambda, вычислите среднюю оценку всех студентов из исходного списка. Результат округлите до двух знаков после запятой.

```python
from functools import reduce
average_grade = round(reduce(lambda x, y: x + y['grade'], students, 0) / len(students),2)
average_grade
```

# Задача 2. Функциональная обработка данных в Pandas

Дан DataFrame с информацией о продажах:

```python
import pandas as pd
import numpy as np
np.random.seed(42)
data = {
'product': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
'price': [100, 150, 100, 200, 150, 100, 200, 150],
'quantity': np.random.randint(1, 10, size=8)
}
df = pd.DataFrame(data)
print(df)
```

Используя map для Series, создайте новый столбец 'revenue', который будет равен price * quantity.

```python
df['revenue'] = df['price'] * df['quantity']
print(df)
```

Сгруппируйте данные по продукту и вычислите общую выручку для каждого продукта.

```python
print("Выручка по продуктам:")
df.groupby('product')['revenue'].sum()
```

Используя apply с lambda-функцией, отфильтруйте строки, где количество проданных единиц (quantity) больше среднего значения quantity по всему DataFrame. Выведите получившийся DataFrame.

```python
print(f'Продажи с quantity > среднего ({round(df['quantity'].mean(),1)}):')

filtered_df_quantity = df[df['quantity'].apply(lambda x: x > df['quantity'].mean())]
filtered_df_quantity
```

# Задача 4. Декоратор для измерения времени выполнения
Реализуйте декоратор timer, который:
1. Замеряет время выполнения декорируемой функции
2. Выводит на экран сообщение: "Функция {имя_функции} выполнилась за {время:.4f} сек."
3. Возвращает результат выполнения функции
4. Корректно работает с любыми функциями (любое количество аргументов)
5. Сохраняет имя (__name__) и документацию (__doc__) исходной функции

Примечание: п.5 реализовать вручную или использовать functools.wrap, который копирует метаданные (__name__, __doc__) из исходной функции в функцию-обёртку.
Примените этот декоратор к функции, которая вычисляет сумму чисел от 1 до 1_000_000, и выведите результат.

```python
import time
from functools import wraps

def timer(func):
 @wraps(func)
 def wrapper(*args, **kwargs):
  start_time = time.time()
  result = func(*args, **kwargs)
  time.sleep(2.9) # Дополнительная пауза для выполниения снимка
  end_time = time.time()
  execution_time = end_time - start_time
  print(f"Функция {func.__name__} ({func.__doc__}) выполнилась за {execution_time:.4f} сек.")
  return result
 return wrapper

@timer
def sum_numbers(n):
 """Вычисляет сумму чисел от 1 до n."""
 return sum(range(1, n + 1))

# Применяем декоратор к функции и вызываем её
result = sum_numbers(1000000)
print(f"Результат вычисления: {result}")
```
