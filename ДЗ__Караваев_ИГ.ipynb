```python
import time
from concurrent.futures import ThreadPoolExecutor, as_completed

def process_number(number):
 """
 Функция обработки числа: умножает число на 2 и имитирует задержку 0.2 секунды.
 """
 time.sleep(0.2)
 return number * 2

def main():
 # Шаг 1: Загрузка данных из файла
 with open('test_list_numbers.Txt', 'r', encoding='utf-8') as file:
  content = file.read()
 
 # Преобразуем строку в список списков (предполагается, что файл содержит валидный Python-код)
  try:
   number_lists = eval(content)
  except Exception as e:
   print(f"Ошибка при чтении файла: {e}")
  return
 
 # Шаг 2: Инициализация ThreadPoolExecutor
 executor = ThreadPoolExecutor(max_workers=10)
 futures = []
 
 # Шаг 3: Отправка задач в пул потоков
 for lst in number_lists:
   for number in lst:
    future = executor.submit(process_number, number)
   futures.append(future)
 
 # Шаг 4: Получение результатов по мере завершения задач
 results = []
 for future in as_completed(futures):
   try:
    result = future.result()
    results.append(result)
   except Exception as e:
    print(f"Задача вызвала исключение: {e}")
 
 # Шаг 5: Находим список, который был обработан быстрее остальных
 # Поскольку все числа обрабатываются параллельно, «быстрее» — это просто порядок завершения
 # Но если нужно найти сумму из самого первого завершённого списка, то:
 # В данном случае мы не храним информацию о принадлежности к списку, поэтому считаем, что
 # все результаты относятся к одному общему списку (по условию задачи).
 
 total_sum = sum(results)
 print(f"Сумма обработанных чисел: {total_sum}")
 
 # Завершаем работу пула
 executor.sutdown()

if __name__ == '__main__':
 main()
```
